        <div id="backLink" onclick="window.location.href='index.html'">Back to Game</div>
    </div>

    <script>
        let score = 0;
        let workers = 0;
        const workerCost = 50;

        const buyWorkerBtn = document.getElementById('buyWorker');

        // Simulating score retrieval from localStorage or a backend
        function getScore() {
            return score;
        }

        function updateShop() {
            buyWorkerBtn.disabled = getScore() < workerCost;
            document.getElementById('workerCount').innerText = workers;
        }

        buyWorkerBtn.addEventListener('click', () => {
            if (getScore() >= workerCost) {
                score -= workerCost;
                workers++;
                updateShop();
            }
        });

        updateShop();
    </script>
