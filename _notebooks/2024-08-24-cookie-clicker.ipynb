    <script>
        let score = 0;
        let workers = 0;
        const workerCost = 50;

        const clickSound = document.getElementById('clickSound');

        document.getElementById('cookie').addEventListener('click', () => {
            score++;
            updateScore();
            clickSound.play();
        });

        function updateScore() {
            document.getElementById('score').innerText = `Score: ${score}`;
            document.getElementById('workerCount').innerText = `Workers: ${workers}`;
            updatePlate();
        }

        function updatePlate() {
            const plate = document.getElementById('plate');
            plate.innerHTML = ''; // Clear existing cookies
            for (let i = 0; i < score; i++) {
                const cookieElement = document.createElement('div');
                cookieElement.className = 'cookieOnPlate';
                plate.appendChild(cookieElement);
            }
        }

        function addWorker() {
            if (score >= workerCost) {
                score -= workerCost;
                workers++;
                updateScore();
            }
        }

        setInterval(() => {
            score += workers;
            updateScore();
        }, 1000);
    </script>
